```python
import pandas as pd
df=pd.read_csv("intern/input/main.csv")
df.head()
```

```python
df.set_index("SKU", inplace = True)
filt=df[df['COUNTRY'].str.contains("USA")]
filt
```

```python
filt.to_csv('filteredCountry.csv')
```

```python
print(df['PRICE'].dtypes)
```

```python
df1=pd.read_csv("filteredCountry.csv")
df1.head()
df1["PRICE"]=df1['PRICE'].str.replace("$","")
df1["PRICE"]=pd.to_numeric(df1["PRICE"],errors = 'coerce')
df1.head()
print(df1['PRICE'].dtypes)
```

```python
df2=df1.groupby('SKU').apply(lambda df: df.nsmallest(2, 'PRICE'))
df2[['PRICE']]
```
